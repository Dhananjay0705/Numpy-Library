**ARRAY FUNCTIONS**

```python
import numpy as np
```

```python
a1 = np.random.random((3,3))
a1 = np.round(a1*100)
a1
```

```python
## max/min/sum/prod
## 0--> col , 1 ---> raws
np.max(a1,axis = 1) ## for rows
```

```python
## mean/median/std/var
np.mean(a1,axis = 1)
np.median(a1,axis = 1)
np.std(a1,axis = 1)
np.var(a1,axis = 1)
```

```python
## trignometric Functions
np.sin(a1)
```

```python
## dot product
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(12,24).reshape(4,3)
np.dot(a2,a3)
```

```python
## Log & Exponents
np.log(a1)
np.exp(a1)
```

```python
## round/floor/ceil
np.round(np.random.random((2,3))*100)
np.floor(np.random.random((2,3))*100) ## floor use for back roundoff
np.ceil(np.random.random((2,3))*100) ## ceil use for forward next integer
```
