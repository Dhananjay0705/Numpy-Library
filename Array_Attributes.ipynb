**ARRAY ATTRIBUTES**

```python
import numpy as np
```

```python
a1 = np.arange(10)
a2 = np.arange(12,dtype = float).reshape(3,4)
a3 = np.arange(8).reshape(2,2,2)
```

```python
## ndim (number of dimension)
a1.ndim
```

```python
## Shape (give information about rows & coloumns)
a3.shape
```

```python
## Size (number of items)
a2.size
```

```python
## itemsize (It shows how much space it is taking in memory in bits)
a1.itemsize
```

```python
## dtype
print(a1.dtype)
print(a2.dtype)
print(a3.dtype)
```
