Creating Numpy Array


```python
import numpy as np
a = np.array([1,2,3])
print(a) ## Result of a is vector
print(type(a)) ## datatype of a
```

```python
## 2D Array
b = np.array([[1,2,3],[4,5,6]])
print(b) ## Result of b is matrix
```

```python
# 3D Array
c = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print(c)
print(c.shape)
```
