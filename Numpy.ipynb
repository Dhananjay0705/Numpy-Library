Creating Numpy Array


```python
import numpy as np
a = np.array([1,2,3])
print(a) ## Result of a is vector
print(type(a)) ## datatype of a
```

```python
## 2D Array
b = np.array([[1,2,3],[4,5,6]])
print(b) ## Result of b is matrix
```

```python
# 3D Array
c = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print(c) ## result of c is tensor
print(c.shape)
```

```python
## Dtype
np.array([1,2,3],dtype=float)
```

```python
## np.arange
np.arange(1,11,2)
```

```python
## with reshape
np.arange(1,26).reshape(5,5)
```

```python
##np.ones
np.ones((3,4))
```

```python
## np.zeros
np.zeros((3,4))
```

```python
## np.random
np.random.random((3,4))
```

```python
## np.linspace
np.linspace(-10,10,10)
```

```python
## np.identity
np.identity(3)
```
