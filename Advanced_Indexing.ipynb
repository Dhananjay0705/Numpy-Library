**Advanced Indexing**

```python
import numpy as np
```

```python
## Normal Indexing
a = np.arange(24).reshape(6,4)
a
```

```python
a[1,1]
```

```python
## Normal Slicing
a[1:3,1:3]
```

```python
## Fancy Indexing
a[[0,4,5]] ## 0,4,5 are rows in indexing number
a[:,[2,3]] ## 2,3 are columns in indexing number
```

```python
## Boolean Indexing
b = np.random.randint(1,100,24).reshape(6,4)
```

```python
## Find all numbers greater than 50
b > 50
```
