**Iterating**

```python
import numpy as np
```

```python

a1 = np.arange(10)
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(27).reshape(3,3,3)
```

```python
for i in a1:
  print(i)
```

```python
for i in a2:
  print(i)
```

```python
a3
```

```python
for i in np.nditer(a3) :
  print(i)
```
