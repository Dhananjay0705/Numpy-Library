**Iterating**

```python
import numpy as np
```

```python

a1 = np.arange(10)
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(27).reshape(3,3,3)
```

```python
for i in a1:
  print(i)
```

```python
for i in a2:
  print(i)
```

```python
a3
```

```python
for i in np.nditer(a3) :
  print(i)
```

**RESHAPE**

```python
## Transpose
np.transpose(a2) ## transpose the matrix convert rows into columns and columns into rows

##  a2.T is give same output of np.transpose(a2)
```

**Stacking**

```python
## horizontal Stacking ## This is use for merge multiple datasource
a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)
np.hstack((a4,a5)) ## a4,a5 ko horizontal stack karna
```

```python
## Vertical Stacking ## This is use for merge multiple datasource
np.vstack((a4,a5))

## the rule for stacking is shape same
```

**Splitting**

```python
## Horizontal Splitting ## This is use for split thr data from the datasource
a4
```

```python
np.hsplit(a4,2) ## It has to be a equal division
```

```python
## Vertical Splitting ## This is use for split thr data from the datasource
np.vsplit(a5,3)
```
